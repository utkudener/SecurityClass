1. Read the text file `wikipedia_cybersecurity.txt` 
2. Build a function `count_occurrence` that given a certain letter of the English alphabet, returns its occurrence in the text
3. Build a function `compute_distribution` that given a text, returns the English letters distribution (frequency) contained in a suitable data structure
4. Visualize the letters distribution by means of the most appropriate plot
5. Save the data structure as a `.pkl` ("pickle") file in the data folder


```python
import string
import matplotlib.pyplot as plt 
import pickle 
import os 

#Reading the text file
file_path = "wikipedia_cybersecurity.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text_data = file.read()

#Function of occurrence counter
def counter_occ(letter, text):
    return text.lower().count(letter.lower())

def computer_distribution(text):
    distribution = {}
    text_lower = text.lower()

    for char in string.ascii_lowercase:
        distribution[char] = text_lower.count(char)

    return distribution
letter_dist = computer_distribution(text_data)

letters = list(letter_dist.keys())
counts = list(letter_dist.values())

plt.figure(figsize=(10,5))
plt.bar(letters, counts, color = "skyblue", edgecolor="black")
plt.title("Letter Distribution")
plt.xlabel("Letters")
plt.ylabel("Frequency")

plt.show()

os.makedirs("notebooks", exist_ok=True)


with open("letter_distribution.pkl",  "wb") as pkl_file:
    pickle.dump(letter_dist, pkl_file)
```
